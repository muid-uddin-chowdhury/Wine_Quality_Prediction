```python
X = wine_dataset.drop('quality', axis = 1)
print(X)
```

**Label Binarize, binarize the quality columns**

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value >= 7 else 0) #lembda is replacing 0-6 in 0 and 7-8 in 0
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state = 4)
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_predict = model.predict(X_train)
predict_train = accuracy_score(X_train_predict, Y_train)
print(predict_train)
```

```python
X_test_predict = model.predict(X_test)
predict_test = accuracy_score(X_test_predict, Y_test)
print(predict_test)
```

```python
input_data = (7.3,0.65,0.0,1.2,0.065,15.0,21.0,0.9946,3.39,0.47,10.0)

input_data_as_numpy_array = np.asarray(input_data)

#reshape just give predict about a particular row
input_data_reshape = input_data_as_numpy_array.reshape(1, -1) #one single row

prediction = model.predict(input_data_reshape) #input_data_reshape putting into random forrest classifier.
print(prediction)
```

```python
if (prediction[0] == 1):
  print('It is a Good wine.')
else:
  print('It is a Bad wine')
```
